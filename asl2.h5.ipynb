```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Dropout
from keras.layers import Flatten
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
x_train = train_datagen.flow_from_directory(r'C:\Users\Lenovo\Desktop\final project\a to e', target_size=(64,64), batch_size=30, class_mode='categorical')
```

```python
x_train.class_indices
```

```python
x_test = test_datagen.flow_from_directory(r'C:\Users\Lenovo\Desktop\final project\OneDrive_1_8-12-2020\a to e', target_size=(64,64), batch_size=30, class_mode='categorical')
```

```python
x_train.image_shape
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32, (3,3), input_shape=(64,64,3), activation = 'relu'))
#no. of feature detectors, size of featuredetector, image size, activation function
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Convolution2D(64, (3,3), input_shape=(64,64,3), activation = 'relu'))
#no. of feature detectors, size of featuredetector, image size, activation function
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Convolution2D(128, (3,3), input_shape=(64,64,3), activation = 'relu'))
#no. of feature detectors, size of featuredetector, image size, activation function
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=512, activation='relu'))
```

```python
model.add(Dense(units=1024, activation='relu'))
```

```python
model.add(Dense(units=5, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit_generator(x_train, steps_per_epoch=500, epochs=10, validation_data=x_test, validation_steps=40)
#steps_per_epoch = no. of train images//batch size
```

```python
model.save('asl2.h5')
```

```python
from keras.models import load_model         # This is used to load the model
from keras.preprocessing import image       # This is used to preprocess the image.
import numpy as np
import cv2
```

```python
model = load_model('asl2.h5')       # load the model
```

```python
video = cv2.VideoCapture(0)        # used to capture an image from the video, 0 means it will trigger the webcam.
name = ['A','B','C','D','E']      # specify the strings that you have in your class indices. 
while video.isOpened():     # infinite loop to read the images from the camera
    success, frame = video.read()      # read the images from the camera. video.read() returns two parameters, success - if it is a success or not, and frame - the image will be stored in frame
    # if you print the frame, you will get all the pixel values displayed.
    cv2.namedWindow('frame',cv2.WINDOW_NORMAL)   # create a window in which you can see the predictions, instead of seeing it on the terminal. 
    # Here 'frame' is the name of the window, cv2.WINDOW_NORMAL is the size of the window. You can can give auto size.
    cv2.resizeWindow('frame',(600,600))      # display the 'frame' and the size of the window should be 600 by 600 
    frame1 = cv2.resize(frame, (64,64))   # Since our image is present in 'frame', we will have to resize it.
    # (64,64) specifies the size of the image. frame1 will contain the resized image of what your camera has captured.
    x = image.img_to_array(frame1)      # convert the image to an array
    x = np.expand_dims(x, axis=0)    # Add an extra dimension for the convolution layer.
    pred = model.predict_classes(x)   # predict the class that the image belongs to. This returns a list. pred[0] will have the output.
    print(pred)    # display it on the console
    # putText displays the output on the window screen
    # frame specifies the image, and then text specifies the text to be displayed
    # org = (10,40) specifies where the text has to originate from. It starts from (0,0) from the top
    # fontFace specifies the font that we want to print it in
    # fontScale specifies the size of the font. 
    # color specifies the color of the font.
    cv2.putText(frame, text="This belongs to "+str(name[pred[0]]), org = (10,40), fontFace=cv2.FONT_HERSHEY_PLAIN, fontScale = 1, color=(0,255,255))
    cv2.imshow('frame',frame)    # Show the image / frame. 'frame' specifies the frame.
    if (cv2.waitKey(1) & 0xFF == ord('q')):    # condition to exit the infinite loop. waitKey is used to wait till any key is entered.
        # ord('q') just specifies if the key is q.
        break
        
video.release()   # release the video, just like closing the file.
cv2.destroyAllWindows()     # destroy all the windows 
```
