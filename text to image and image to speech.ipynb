```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
import gtts
from playsound import playsound
```

```python
import pyttsx3
engine = pyttsx3.init()
```

```python
model = load_model('asl2.h5')
```

```python
import cv2
```

```python
video = cv2.VideoCapture(0)
name = ['A', 'B', 'C', 'D', 'E']
while(True):
    success, frame = video.read()
    cv2.namedWindow('frame', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('frame',(800,800))
    frame1 = cv2.resize(frame,(64,64))
    x = image.img_to_array(frame1)
    x = np.expand_dims(x, axis=0)
    pred = model.predict_classes(x)
    text = name[pred[0]]
    # play the speech
    print(pred)
    cv2.putText(frame, text = "This belongs to "+str(name[pred[0]]), org=(10,40), fontFace=cv2.FONT_HERSHEY_PLAIN, fontScale=1, color=(1,0,0))
    cv2.imshow('frame',frame)
    engine.say(text)
    engine.runAndWait()
    if(cv2.waitKey(1) & 0xFF == ord('q')):
        break
        
video.release()
cv2.destroyAllWindows()
```
